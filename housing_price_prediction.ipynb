# using simple linear regression to predict housing price per square fit.
variable x = price
variable y = square fit area of house

1) Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

2) Loading dataset


```python
dataset = pd.read_csv("Housing.csv")
x = dataset.iloc[:, [5]].values
y =dataset.iloc[:, 2].values
```

3) Spliting data into training and test set


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

4) predicting y values

```python
y_pred = regressor.predict(x_test)
```

5) Visualizing training set


```python
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, regressor.predict(x_train), color = "blue")
plt.title("training set visualization")
plt.xlabel("square fit")
plt.ylabel("price")
plt.show()
```

5) Visualizing test set


```python
plt.scatter(x_test, y_test, color = "orange")
plt.plot(x_train, regressor.predict(x_train), color = "blue")
plt.title("test set visualization")
plt.xlabel("square fit")
plt.ylabel("price")
plt.show()
```
